```
#We only consider attributes of an image which provide information called points of interest.
#Edges and corners are two common types of points of interest.
#An edge is boundary where pixel intensity rapidly changes and corner is intersection of two edges
#We extract points of interest using scikit-learn
```

```
import numpy as np
from skimage.feature import corner_harris, corner_peaks
from skimage.color import rgb2gray
import matplotlib.pyplot as plt
import skimage.io as io
from skimage.exposure import equalize_hist
def show_corners(corners,image):
  #The figure() function in pyplot module of matplotlib library is used to create a new figure.
  fig=plt.figure()
  plt.gray() #Set the colormap to "gray".
  plt.imshow(image) 
  y_corner,x_corner=zip(*corners) #unzipping a sequence
  plt.plot(x_corner,y_corner,'or')
  plt.xlim(0,image.shape[1]) #x axis limits and start
  plt.ylim(image.shape[0],0) #y axis limit and start
  fig.set_size_inches(np.array(fig.get_size_inches())*1.5) #The set_size_inches() method figure 
  #module of matplotlib library is used to set the figure size in inches.
  plt.show()
madrid=io.imread('/content/bridge-coast-golden-gate-bridge.jpg')
madrid=equalize_hist(rgb2gray(madrid)) #Return image after histogram equalization.
#corner_peaks: Find corners in corner measure response image.
#corner_harris: Compute Harris corner measure response image.
#This corner detector uses information from the auto-correlation matrix
corners=corner_peaks(corner_harris(madrid),min_distance=2) 
show_corners(corners,madrid)
```

```
#SIFT(Scale Invariant Feature Transform) AND SURF(Speeded Up Robust Features)
#SIFT is a method for extracting features from an image which is less sensitive to rotation,scaling,
#and illumination of image than the extracted methods we have previously worked in.
#SURF is another method of extracting interesting points of an image and creating descriptions that are
#invariant of image's scale, orientation,and illumination. SURF can be computed more quickly than SIFT
#and it is more effective at at recognizing features accross images.
#We will use SURF in the below example.
#just like extracted points of interest, the SURF are only the first step in creating a feature representation
#that could be used in ML task. Different SURF will be extracted for each instance in training set. 
```

```
import mahotas as mh
from mahotas.features import surf
image=mh.imread('/content/real_madrid_metal_logo-t2.jpg',as_grey=True)
print('The first SURF descriptor:',surf.surf(image)[0])
print('Extracted %s SURF descriptors'%len(surf.surf(image)))
```
